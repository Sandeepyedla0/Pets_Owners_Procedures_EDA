# 8Q What percentage of cities have more male pets than female pets?


```python
## Grouping data with Cities having more males
grouped_data = owners_pets_joined.groupby(['City', 'Gender']).size().reset_index(name='Count').sort_values(by = 'City', ascending = True)
```

```python
grouped_data.head()
```

```python
# creating a pivot table and having male and female filled with repective counts
pivoted_data = grouped_data.pivot(index='City', columns='Gender', values='Count')
pivoted_data[['female','male']] = pivoted_data[['female','male']].astype('Int64')
```

```python
pivoted_data.head()
```

```python
# filling the NA values with 0 for better representation 
pivoted_data.fillna(0,inplace=True)
```

```python
pivoted_data.head()
```

```python
# creating additional column to mark '1' if male > female else label as "0"
pivoted_data['Male_greaterto_Female'] = np.where(pivoted_data['male']> pivoted_data['female'], 1, 0)
```

```python
pivoted_data.head()
```

```python
# summing up the Male_greaterto_Female column to get how many cities have male pets greater than female
pivoted_data['Male_greaterto_Female'].sum() 
```

```python
# getting count of unique city to calculate percentage
len(owners_pets_joined['City'].unique())
```

```python
## Calculating the percentage
percentage_MgF_cities = round(pivoted_data['Male_greaterto_Female'].sum() / len(owners_pets_joined['City'].unique()) *100,2)
print("Cities having males greater than female: {:.2f}%".format(percentage_MgF_cities))
```

## Ans: Cities having males greater than female: 57.50%

# 9Q Which city's pet sample is made up of exactly 70% dogs?

```python
# Group by data with respect to city and kind and count the occurances
city_dog = owners_pets_joined.groupby(['City', 'Kind']).size().reset_index(name='Count').sort_values(by = 'City', ascending = True)
```

```python

city_dog.head()
```

```python
# Pivoting the data with column kind to get the each pet count in cities
city_dog = city_dog.pivot(index = 'City',columns = 'Kind', values = 'Count' )
```

```python
city_dog.fillna(0,inplace=True)
city_dog.head()
```

```python
# create a column to have total pets in each city to calculate the dogs percentage
city_dog['Total_Pets'] = city_dog.sum(axis=1)
city_dog.head()
```

```python
# Calculate the percentage of dogs and total pets in city
city_dog['Percentage_Dogs'] = (city_dog['Dog'] / city_dog['Total_Pets'] * 100)
# using round off funtion to round the numerical to 2 decimals
city_dog['Percentage_Dogs'] = city_dog['Percentage_Dogs'].round(2)
```

```python
# Filtering the column 'Percentage_Dogs' with 70 percentage
city_dog[city_dog['Percentage_Dogs']==70.00]
```

## Ans Grand Rapids city's pet sample is made up of exactly 70% dogs
